```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 800)
```

```python
zipcode = pd.read_csv('./zipcode.csv')
zipcode = zipcode.sort_values(by='ZIPCODE').drop_duplicates(subset='ZIPCODE')
```

```python
business = pd.read_csv('./Legally_Operating_Businesses.csv')
```

```python
business.shape
```

```python
business = business.apply(lambda x: x.astype(str).str.lower())
```

```python
# filter out none-NY businesses
boros = ['brooklyn', 'manhattan', 'queens', 'staten island', 'bronx', 'new york']
business_ny = business[business['Address City'].isin(boros)]
business_ny.shape
```

```python
# get valid zipcodes
business_ny = business_ny[business_ny['Address ZIP'].isin(zipcode['ZIPCODE'].astype(str))]
```

```python
pd.get_dummies(business_ny['License Category']).shape
```

```python
business_ny_dummies = pd.concat([business_ny, pd.get_dummies(business_ny['License Category'])], axis=1)
business_ny_zip = business_ny_dummies.groupby('Address ZIP', as_index=False)
```

```python
business_summary = pd.DataFrame(business_ny_zip.count().as_matrix()[:,:2], columns=['zipcode','count'])
business_summary = pd.concat([business_summary, business_ny_zip.sum().drop('Address ZIP', axis=1)], axis=1)
business_summary.head()
```

```python
business_summary.to_csv('./business_summary.csv')
```

```python
restaurants = pd.read_csv('./restaurant_summary.csv')
restaurants
```

```python
zipcode.dtypes
```

```python
restaurants = restaurants[['ZIPCODE','restaurant_amount']]
```

```python
zipcode = zipcode[['ZIPCODE','AREA','POPULATION']]
```

```python
res_zip = restaurants.set_index('ZIPCODE').join(zipcode.set_index('ZIPCODE'), how='inner', lsuffix='_r', rsuffix='_z')
res_zip.head()
```

```python
res_zip['index'] = np.log(res_zip['restaurant_amount']/res_zip['AREA']*2000000)
res_zip.sort_values(by='index',ascending=False).head()
```

```python
res_zip.to_csv('./restaurant_index.csv')
```
